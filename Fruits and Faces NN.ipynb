```python
# Importing the necessary libraries
import tensorflow as tf
import numpy as np 
import matplotlib.pyplot as plt
import os
from tensorflow import keras
```

### Loading the images to train and test 

In order to create the model for the NN it is necessary to have a Database, in this case the Database for both was downloaded from *Kaggle*

```python
# Creating a data generator
datagen = keras.preprocessing.image.ImageDataGenerator(dtype = 'uint8')
```

```python
# Loading and iterating through the training dataset
train_it = datagen.flow_from_directory('fruits-360_dataset/fruits-360/Training/', class_mode = 'binary', batch_size = 64)
# Loading and iterating through the testing dataset
test_it = datagen.flow_from_directory('fruits-360_dataset/fruits-360/Test/', class_mode = 'binary', batch_size = 64)
```

```python
# Reading and showing different images
plt.figure(figsize = (5,5))
for i in range(10):
    plt.subplot(5,5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(plt.imread(train_it.filepaths[i * 100]))
```
